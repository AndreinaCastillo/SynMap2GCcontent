```python
#SynMap2GCcontent
##Step 1 Importing FASTA files

####Import library
import os

####Global variables 
header = []
nucleotides = []
all_fractions = []

####Paths
#my_file = open(os.path.expanduser('Plasmodium_chabaudi_chabaudi_strain_AS.faa'))
#my_file = open(os.path.expanduser('Plasmodium_ovale_curtisi.faa'))
my_file = open(os.path.expanduser('minitestdata.fasta'))

file_contents = my_file.read()
file_contents_lower = file_contents.lower()                                                    #Lowercases all strings in file
file_contents_lower_split = file_contents_lower.split()                                        #Split the whole file string into words
                                                                       

for i in file_contents_lower_split:                                                            #Divides words into two lists: one with the simbol ">" and the other one without it
    if ">" in i:
        header.append(i)
    else:
        nucleotides.append(i)

dictio = dict(zip(header, nucleotides))                                                        #Makes a dictionary using the two created lists: header and nucleotides


for seq in nucleotides:                                                                        #Loops for every sequence in the nucleotide list
    seq_frac = []
    count_a = seq.count('a')                                                                   #Counts the number of each nucleotide per sequence/chromosome in nucleoide list
    count_t = seq.count('t')
    count_c = seq.count('c')
    count_g = seq.count('g')
    count_n = seq.count('n')
    count_x = seq.count('x')

    frac_a = (float(count_a/len(seq)))*100                                                     #Divides the counted nucleotide by the sequence lenght and multiplies it by 100 (created percentage)
    frac_c = (float(count_c/len(seq)))*100
    frac_t = (float(count_t/len(seq)))*100
    frac_g = (float(count_g/len(seq)))*100
    frac_n = (float(count_n/len(seq)))*100
    frac_x = (float(count_x/len(seq)))*100

    seq_frac.append(round(frac_a,2))                                                           #Appends each count into a list unique per sequence/chromosome in nucleotides list. It rounds the decimal point to two
    seq_frac.append(round(frac_c,2))
    seq_frac.append(round(frac_t,2))
    seq_frac.append(round(frac_g,2))
    seq_frac.append(round(frac_n,2))
    seq_frac.append(round(frac_x,2))

    all_fractions.append(seq_frac)                                                              #Appends all nucleotide fractions/percentages to a single list

dictio_fractions = dict(zip(header, all_fractions))                                             #Combines sequence/chromosome headers and the calculated nucleotide fractions into a dictionary
print(dictio_fractions)
```
