```python
#PlotlyinputGCcontent
##Step 1 Importing FASTA files


#####Credentials
import plotly.tools as tls
tls.set_credentials_file(username='AndreinaCastillo', api_key='i7hoddkosa')

####Import library
import os                                                                                      #Module to read files
# import plotly.plotly as py
# import plotly.graph_objs as go
import numpy as np
from plotly import __version__
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
from plotly.graph_objs import Bar, Scatter, Figure, Layout
init_notebook_mode(connected=True)

####Global variables 
header = []
nucleotides = []

#File input/path

#my_file = open(os.path.expanduser('Plasmodium_chabaudi_chabaudi_strain_AS.faa'))
my_file = open(os.path.expanduser('Plasmodium__test.faa'))
file_contents = my_file.read()
file_contents_lower = file_contents.lower()                                                    #Lowercases all strings in file
file_contents_lower_split = file_contents_lower.split()                                        #Split the whole file string into words
                                                                       

##Step 2 Separating the file from a  single string into two lists

for i in file_contents_lower_split:                                                            #Divides words into two lists: one with the simbol ">" and the other one without it
    if ">" in i:
        header.append(i)
    else:
        nucleotides.append(i)

        
##Step 3 Reading the nucleotides list and counting the type of nucleotides (counter) 
##and the lenght per sequence (positioner)

def counter(seq):
#Loops on every nucleotide per sequence. Then, per each time if find a G or C it add a 1 in that position to 
#the count list, for A, T, N and X adds a 0
    count = []
    for nuc in seq:                                                          
        if nuc == 'a':
            count.append(0)
        if nuc == 't':
            count.append(0)
        if nuc == 'c':
            count.append(1)
        if nuc == 'g':
            count.append(1)
        if nuc == 'n':
            count.append(0)
        if nuc == 'x':
            count.append(0)
        count_array = np.asarray(count)
    return count_array

def positioner(seq):
#Loops for every nucleotide per sequence, counts the number of positions and adds them to the position list
    l = 0
    position = []
    for l in range(0,len(seq)):                                                              
        l = l+1
        position.append(l)           
        position_array = np.asarray(position)
    return position_array


def plotly(nucleo):
#This function builds a plot using the nucleotide positions and G,C,T,A values,
#and allows to zoom in in a desired areas
    for seq in nucleotides:
        print(seq)
        pos_arr = counter(seq)
        cnt_arr = positioner(seq)
        trace = Scatter(
        y = pos_arr,
        x = cnt_arr)
        
        data = [trace]   
        print (data)
        #iplot(data, filename='basic-line')
    return


# def plotly(nucleo):
# This function builds a plot using the nucleotide positions and G,C,T,A values 
# BUT ALSO outputs graph to a new window and allows to zoom in in a desired area
#     for seq in nucleotides:
#         #print(seq)
#         pos_arr = counter(seq)
#         #print (pos_arr)
#         cnt_arr = positioner(seq)
#         #print (cnt_arr)
#         trace = Scatter(
#         y = pos_arr,
#         x = cnt_arr)
        
#         data = [trace]
        
#         layout = Layout(
#         autosize=False,
#         width=500,
#         height=1,
#         paper_bgcolor='#7f7f7f',
#         plot_bgcolor='#c7c7c7'
#         )
#         fig = Figure(data=data, layout=layout)
#         plot_url = plot(fig, filename='size-margins.html')
#     return


#counter(nucleotides)
#positioner(nucleotides)
plotly(nucleotides)
```
